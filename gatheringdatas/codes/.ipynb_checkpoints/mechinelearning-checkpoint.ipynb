### 벡터 토큰화(단어 숫자화)

```python
import pandas as pd
df_review_positive = pd.read_csv('../datasets/reviews_data/all_preprocess_positive.csv')
df_review_negative = pd.read_csv('../datasets/reviews_data/all_preprocess_negative.csv') 
```

```python
df_review_negative = df_review_negative[condition_negative]
```

```python
comments_positive = df_review_positive['replaced_review'].tolist()
comments_positive
```

```python
comments_positive = [comment for comment in comments_positive if not pd.isna(comment)]
```

```python
comments_negative = df_review_negative['replaced_review'].tolist()
comments_negative
```

```python
comments_negative = [comment for comment in comments_negative if not pd.isna(comment)]
```

### 형태소 분석기 통한 품사 분류

```python
from mecab import MeCab
mecab = MeCab()
```

```python
comments_list
```

```python
mecab.pos(comments_list[8600])
```

```python
tokenized_comments = list()
for comment in comments_list :
    comment_morphs = mecab.morphs(comment)
    # print(' '.join(comment_morphs))
    # 추가로 불용어 처리 필요
    tokenized_comments.append(comment_morphs)
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
comments = list()
for token in tokenized_comments:
    comments.append(' '.join(token))
# 리스트 안의 리스트를 한 문장으로 결합
```

```python
comments 
```

```python
countVectorizer = CountVectorizer()
countVectorizer.fit(comments)
```

```python
countVectorizer.vocabulary_
```

```python
tfidfVectorizer = TfidfVectorizer(use_idf=True)
tfidfVectorizer.fit(comments)
```

```python
tfidfVectorizer.idf_
```

```python
feature = tfidfVectorizer.transform([comments[9]]) # 학습 시 2차원을 이용했으므로 변환도 2차원 적용
feature.toarray() # vocabulary_ 갯수와 사이즈가 같음.
```

```python
features = tfidfVectorizer.transform(comments) # 댓글 전체 벡터화
```

### 목표 변수 생성

```python
target = ['긍정'] * len(comments_positive)  + ['부정'] * len(comments_negative)
# target
```

### 머신러닝 학습

```python
from sklearn.svm import SVC
```

```python
model = SVC()
model.fit(features.toarray(), target)
```

### 머신러닝 통한 예측

```python
comment = '코로나19의 여파로 전 세계 교사들과 학생들 모두 혼란스러운 시기를 보내고 있습니다.어려움 속에서도 학생들을 위해 안간힘을 쓰고 있는 모든 선생님들에게, 응원과 박수를 함께 보냅니다.'
```

```python
# 일단 이거 쪼갠다
tokenized_comment =mecab.morphs(comment)
```

```python
vectorized_comment = tfidfVectorizer.transform([' '.join(tokenized_comment)])
```

```python
vectorized_comment.toarray() # vacabrary에 일치되는것만 변환되기때문
```

```python
model.predict(vectorized_comment.toarray()) #이렇게 되면 comment의 문장은 긍정에 가깝다는 예측이 됨. 
```
