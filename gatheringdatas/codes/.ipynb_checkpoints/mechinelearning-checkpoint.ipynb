### 형태소 분석기 통한 품사 분류

```python
from mecab import MeCab
mecab = MeCab()
```

```python
comments_list
```

```python
# len(comments_list)
```

```python
mecab.pos(comments_list[8600])
```

```python
tokenized_comments = list()
for comment in comments_list :
    comment_morphs = mecab.morphs(comment)
    # print(' '.join(comment_morphs))
    # 추가로 불용어 처리 필요
    tokenized_comments.append(comment_morphs)
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
comments = list()
for token in tokenized_comments:
    comments.append(' '.join(token))
# 리스트 안의 리스트를 한 문장으로 결합
```

```python
comments 
```

```python
# 단어의 출현 빈도를 기반으로 문서를 벡터로 변환
countVectorizer = CountVectorizer()
countVectorizer.fit(comments)
```

```python
countVectorizer.vocabulary_
```

```python
len(countVectorizer.vocabulary_)
```

```python
tfidfVectorizer = TfidfVectorizer(use_idf=True)
tfidfVectorizer.fit(comments)
```

```python
# 문서 간에 고유한 단어의 중요성을 반영
tfidfVectorizer.idf_
```

```python
feature = tfidfVectorizer.transform(comments) # 학습 시 2차원을 이용했으므로 변환도 2차원 적용
```

```python
feature.toarray() # vocabulary_ 갯수와 사이즈가 같음.
```

```python
import numpy as np
```

```python
# 2차원 배열 
features = tfidfVectorizer.transform(comments) # 댓글 전체 벡터화
features
```

### 목표 변수 생성

```python
# 1차원 배열
target = ['긍정'] * len(comments_positive)  + ['부정'] * len(comments_negative)
target
```

#### 정형화 

```python
from sklearn.model_selection import train_test_split
# features와 target을 NumPy 배열로 변환
features_np = np.array(features.toarray()) # features를 NumPy 배열로 변환
target_np = np.array(target) # target을 NumPy 배열로 변환
```

```python
# train_test_split을 사용하여 데이터 분할
features_train, features_test, target_train, target_test = train_test_split(features_np, target_np, random_state=111)
```

```python
# 각 데이터의 형태 확인
print(features_train.shape, target_train.shape, features_test.shape, target_test.shape)
```

### 머신러닝 학습

```python
from sklearn.svm import SVC
```

```python
# SVC 모델 생성, probability=True 확률 예측 위해 설정함.
model = SVC(probability=True)

# 모델 학습
model.fit(features_train, target_train)
```

```python
# 모델 평가
accuracy = model.score(features_test, target_test)
print("Accuracy:", accuracy)
```

##### 머신러닝 예측(features_train)

```python
model.predict(features_train)
```

```python
# 예측 클래스의 확률얻기 위해 predict_proba 메서드 호출함.
model.predict_proba(features_train)
```

### 평가

```python
target_train_predict = model.predict(features_train)
target_train_predict.shape
```

```python
from sklearn.metrics import accuracy_score # 정확도
```

```python
accuracy_score(target_train, target_train_predict) # 교내 시험
```

```python
target_test_perdict = model.predict(features_test)
target_test_perdict.shape
```

```python
accuracy_score(target_test, target_test_perdict) # 교외 시험
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(target_train, target_train_predict))
```

```python
print(classification_report(target_test, target_test_perdict))
```
