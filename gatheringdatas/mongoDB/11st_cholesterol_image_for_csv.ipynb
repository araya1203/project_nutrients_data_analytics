```python
import pymongo as mg
import pandas as pd
```

```python
mg.__version__
```

```python
client = mg.MongoClient(host='mongodb://localhost:27017')
```

```python
database = client['project_nutrients']
```

```python
collection = database['11st_corelsterolsupplement_images']
```

```python
cursor = collection.find({})
```

```python
list_11stcho = list(cursor)
```

```python
list_11stcho
```

```python
df_11stcho = pd.DataFrame(list_11stcho)
```

```python
df_11stcho
```

```python
# 불필요한 columns 제거 
df_11stcho = df_11stcho.drop(columns = ['_id'])
df_11stcho
```

```python
len(df_11stcho)
```

```python
# 중복 데이터를 제거
df_11stcho_drop_duplicates = df_11stcho.drop_duplicates()
```

```python
len(df_11stcho_drop_duplicates)
```

```python
df_11stcho_drop_duplicates.to_csv('../datasets/cholesterol_data/11st_corelsterolsupplement_images.csv')
```
