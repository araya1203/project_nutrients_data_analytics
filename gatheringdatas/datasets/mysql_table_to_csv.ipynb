```python
# 특정 데이터베이스 선택 (이미 선택된 상태라면 생략 가능)
database_name = 'project_nutrients'  # 선택할 데이터베이스 이름
```

```python
cursor.execute(f"USE {database_name}")
```

```python
# 특정 테이블 선택 (USE 문을 사용하지 않고 직접 테이블을 선택)
table_name = 'dummy_tables_supp'  # 선택할 테이블 이름
cursor.execute(f"SELECT * FROM {table_name}")
```

```python
# 결과를 DataFrame으로 변환
data = cursor.fetchall()  # 모든 행을 가져옴
columns = [desc[0] for desc in cursor.description]  # 열 이름을 가져옴
df = pd.DataFrame(data, columns=columns)
```

```python
df
```

```python
df = df.drop(columns=['COL_A', 'COL_B', 'COL_C', 'COL_D', 'COL_E', 'COL_F'])
```

```python
# 특정 테이블 선택 (USE 문을 사용하지 않고 직접 테이블을 선택)
review_name = 'review'  # 선택할 테이블 이름
cursor.execute(f"SELECT * FROM {review_name}")
```

```python
# 결과를 DataFrame으로 변환
review = cursor.fetchall()  # 모든 행을 가져옴
columns = [desc[0] for desc in cursor.description]  # 열 이름을 가져옴
df_review = pd.DataFrame(review, columns=columns)
```

```python
merged_df = df.merge(df_review[['REVIEW_ID', 'CONTENT']], on='REVIEW_ID', how='inner')
```
