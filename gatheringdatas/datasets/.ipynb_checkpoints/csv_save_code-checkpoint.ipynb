```python
import pymongo as mg
import pandas as pd
```

```python
mg.__version__
```

## 쿠팡 간 영양제 CSV 저장

```python
client = mg.MongoClient(host='mongodb://localhost:27017')
database = client['project_nutrients']
collection = database['coupang_liver_review']
```

```python
cursor = collection.find({})
```

```python
list_coupang_liver = list(cursor)
list_coupang_liver
```

```python
df_coupang_liver = pd.DataFrame(list_coupang_liver)
df_coupang_liver
```

```python
df_coupang_liver.to_csv('./reviews_data/coupang_liver_review.csv')
```
